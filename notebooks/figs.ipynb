```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
per_strategy = 0.4002499 / 28
clustering_time = 2.3905
sampling_time = 0.001902
ranking_time = 0.00213
clustering = [6*per_strategy + clustering_time, 14*per_strategy + clustering_time, 17*per_strategy+clustering_time]
sampling = [6*per_strategy+sampling_time, 6*per_strategy+sampling_time, 6*per_strategy+sampling_time]
ranking_by_semantics = [6*per_strategy+ranking_time, 6*per_strategy+ranking_time, 6*per_strategy+ranking_time]
index = ['Age', 'BMI', 'Glucose']
df = pd.DataFrame({'Clustering': clustering, 'Sampling': sampling, 'Ranking by Semantics': ranking_by_semantics}, index=index)
ax = df.plot.bar(rot=0)
plt.grid(linestyle=":", color='gray', linewidth=2, axis='y')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel('Time (s)', fontsize=14)
plt.xlabel('Feature', fontsize=14)
plt.show()
```

```python

def f1(precision, recall):
    return 2 * (precision * recall) / (precision + recall)

exhaustive = [0.94, 0.63, f1(0.94, 0.63)]
tree = [0.82, 0.89, f1(0.82, 0.89)]
auto_table = [0.43, 0.62, f1(0.43, 0.62)]
index = ['Precision', 'Recall', 'F1']
df = pd.DataFrame({'Exhaustive': exhaustive,
                    'Search Tree': tree, 
                    "Auto-Tables": auto_table}, index=index)
ax = df.plot.bar(rot=0, color={"Exhaustive": "#004D40", "Search Tree": "#1E88E5", "Auto-Tables": "#FFC107"}, edgecolor='black', linewidth=1.5)
plt.grid(linestyle=":", color='gray', linewidth=2, axis='y')
ax.legend(bbox_to_anchor=(1.2, 1.2), ncol=len(df.columns), fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel('Score', fontsize=14)
# save plot
plt.savefig('figs/fig1.png', bbox_inches='tight')
plt.show()
```

```python
def f1(precision, recall):
    return 2 * (precision * recall) / (precision + recall)

optimistic = [0.71, 0.797, f1(0.71, 0.797)]
iterative1 = [0.72, 0.8, f1(0.72, 0.8)]
iterative3 = [0.724, 0.805, f1(0.724, 0.805)]
index = ['Precision', 'Recall', 'F1']
df = pd.DataFrame({'Steiner Optimistic': optimistic,
                    #'Iter (pass=1)': iterative1, 
                    "Steiner Iter (pass=3)": iterative3}, index=index)
ax = df.plot.bar(rot=0, color={"Steiner Optimistic": "#004D40", "Iter (pass=1)": "#1E88E5", "Steiner Iter (pass=3)": "#FFC107"}, edgecolor='black', linewidth=1.5)
plt.grid(linestyle=":", color='gray', linewidth=2, axis='y')
ax.legend(bbox_to_anchor=(1.1, 1.2), ncol=len(df.columns), fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel('Score', fontsize=14)
# save plot
plt.savefig('figs/trees.png', bbox_inches='tight')
plt.show()
```
