```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import sys
ppath = sys.path[0] + '/../'
```

```python
attrs = ['Age', 'BMI', 'Glucose']
dfs = []
for attr in attrs:
    df = pd.read_csv(ppath + f'exp/pima/imputation/{attr}_quality.csv')
    dfs.append(df)
# concat all dataframes
df = pd.concat(dfs)
```

```python
methods = ['cluster_sampling_1_0.5', 'cluster_sampling_2_0.5', 'cluster_sampling_1_0.7', 'random_sampling_0.33', 'random_sampling_0.5']
colors = ['r', 'g', 'b', 'c', 'm']
markers = ['o', 'v', 's', 'D', 'x']
labels = ['Cluster Sampling (k=1, p=0.5)', 'Cluster Sampling (k=2, p=0.5)', 'Cluster Sampling (k=1, p=0.7)', 'Random Sampling (p=0.33)', 'Random Sampling (p=0.5)']
# plot density plot
fig, ax = plt.subplots(1, 1, figsize=(15, 5))
for i, method in enumerate(methods):
    sns.kdeplot(df[df['method'] == method]['avg_dist'].to_numpy(), color=colors[i], label=labels[i])
plt.xlabel('Average distance (0 is the best)')
plt.ylabel('Density')
plt.legend()
#plt.savefig(ppath + 'exp/pima/imputation/error_density_plot.png')
plt.show()
```

```python

def f1(precision, recall):
    return 2 * (precision * recall) / (precision + recall)

exhaustive = [0.94, 0.63, f1(0.94, 0.63)]
tree = [0.82, 0.89, f1(0.82, 0.89)]
auto_table = [0.43, 0.62, f1(0.43, 0.62)]
index = ['Precision', 'Recall', 'F1']
df = pd.DataFrame({'Exhaustive': exhaustive,
                    'Search Tree': tree, 
                    "Auto-Tables": auto_table}, index=index)
ax = df.plot.bar(rot=0, color={"Exhaustive": "#004D40", "Search Tree": "#1E88E5", "Auto-Tables": "#FFC107"}, edgecolor='black', linewidth=1.5)
plt.grid(linestyle=":", color='gray', linewidth=2, axis='y')
ax.legend(bbox_to_anchor=(1.2, 1.2), ncol=len(df.columns), fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel('Score', fontsize=14)
# save plot
plt.savefig('figs/fig1.png', bbox_inches='tight')
plt.show()
```
