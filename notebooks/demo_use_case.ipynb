```python
import pandas as pd
import numpy as np
import sys
import os
import matplotlib.pyplot as plt
import os
import warnings
warnings.filterwarnings('ignore')

# Project path
ppath = sys.path[0] + '/../'
```

```python
df = pd.read_csv(os.path.join(ppath, 'data', 'pima', 'diabetes.csv'))
# drop bmi == 0
age_max = 60
df = df[df['Age'] != 0]
df = df[df['Age'] < age_max] 
col = "Age"

gap = 10
bins = [i for i in range(0, age_max+1, gap)]
labels = [f"{i}-{i+gap}" for i in range(0, age_max, gap)]
#labels[-1] = f"{age_max-gap}-{age_max}"

df[col + '.binned'] = pd.cut(df[col], bins=bins, labels=labels)
#df[col + '.binned'] = df[col + '.binned'].astype('float64')
# calculate the percentage of diabetes for each bin
df['diabetes'] = df['Outcome']
df = df[[col + '.binned', 'diabetes']]
df = df.groupby(col + '.binned').mean()
df = df.reset_index()
# plot bar chart
plt.bar(df[col + '.binned'], df['diabetes'])
plt.xlabel('Age bins', fontsize=14)
plt.ylabel('Percentage of diabetes', fontsize=14)
# rotate x-axis labels
#plt.xticks(rotation=90)
plt.show()
```

```python
df = pd.read_csv(os.path.join(ppath, 'data', 'pima', 'diabetes.csv'))
# drop bmi == 0
age_max = 60
df = df[df['Age'] != 0]
df = df[df['Age'] < age_max] 
col = "Age"

bins = [19, 35, 45, 65]
labels = ['19-35\nYoung adult', '35-45\nAdult', '45-65\nMiddle-aged']
# Plot a histogram, change the bar color to #11c739
df[col] = pd.cut(df[col], bins=bins, labels=labels)
labels, counts = np.unique(df[col], return_counts=True)
plt.bar(labels, counts, align='center', color='#11c739')
plt.gca().set_xticks(labels)


#plt.bar(df[col], bins=bins, edgecolor='black', color='#11c739')
# Add bin labels to the center of the bars
#plt.xticks(bins, labels, fontsize=16)
# Add title
plt.title('Age', fontsize=24)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylabel('Count', fontsize=20)
# save figure
plt.savefig(os.path.join(ppath, 'notebooks/plots', 'age_demo.png'), dpi=300, bbox_inches = "tight")
plt.show()
```
